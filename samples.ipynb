# 1. Ирисы Фишера

классы:
- **setosa** - target=0
- **versicolor** - target=1
- **virginica** - target=2

```python
df = sklearn.datasets.load_iris(as_frame=True).frame
df.head(5)
```

```python
df.describe()
```

```python
sns.pairplot(df, kind='scatter', hue='target');
```

```python
X_all = df['sepal width (cm)'][:].to_numpy()
Y_all = df['sepal length (cm)'][:].to_numpy()
```

```python
colors = ['magenta', 'black', 'orange']
labels = ['setosa', 'versicolor', 'virginica']
labeled = [False] * 3
for i in range(X_all.shape[0]):
    clss = df['target'][i]
    label = None if labeled[clss] else labels[clss]
    plt.scatter(X_all[i], Y_all[i], color=colors[clss], label=label)
    labeled[clss] = True
plt.legend()
plt.show()
```
